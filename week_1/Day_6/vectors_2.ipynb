```python
import numpy as np

# deining the 3 vectors
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])
C = np.array([7, 8, 9])

# Claculating the dot product of A and B

dot_product_AB = np.dot(A, B)
dot_product_BA = np.dot(B, A)


#checking the commutative property: A.B = B.A

print("Dot product of A and B:", dot_product_AB)
print("Dot product of B and A:", dot_product_BA)

if dot_product_AB == dot_product_BA:
    print("The dot product is commutative: A.B = B.A") 
else:    print("The dot product is not commutative: A.B != B.A")
```

```python
# checking the distrubutive property: A * (B + C) = A.B + A.C

dot_product_A_BC = np.dot(A, B+C)
dot_product_A_B_C = np.dot(A, B) + np.dot(A, C)

print(dot_product_A_BC)
print("Dot product of A and (B + C):", dot_product_A_BC)


if dot_product_A_BC == dot_product_A_B_C:
    print("The dot product is distributive.")
else:
    print("The dot product is not distributive.")
```
